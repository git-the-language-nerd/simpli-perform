```python
def parse_file_type_1(input_file):
    
    my_dict = OrderedDict()
    
    with open(input_file, 'r') as file:
        sentence_number = 1
        current_dict = {}
        
        for line in file:
            if "Input:" in line:
                current_dict['input'] = line.split("Input:")[1].strip()
            elif "Output:" in line:
                current_dict['output'] = line.split("Output:")[1].strip()
                my_dict[sentence_number] = current_dict
                sentence_number += 1

                # calculate sequence matching ratio to find sentence pairs with no or just trivial changes
                current_dict['Lev_distance'] = Levenshtein.distance(current_dict["input"], current_dict["output"])
                
                current_dict = {}

    

    return my_dict

```
